# Shopify Data Science Challenge

#### 100 sneaker stores sell the similar sneaker and given the transaction data it is required to verified the Average order value

#### Average Order Value (AOV) - the average of total orders made for the sneaker in all those stores 

$ AOV = (N1 Orders + N2 Orders + ..... + N100 Orders) / 5000 $

### inserting the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### importing the data

```python
sneakers = pd.read_csv("2019 Winter Data Science Intern Challenge Data Set - Sheet1.csv")
sneakers
```

#### there are visible difference between the total cost of orders on similar no of items sold which means each store is selling on different price point, let see...

```python
sneakerCost = []

for order_id in sneakers:
    sneakerCost = sneakers["order_amount"] / sneakers["total_items"]

#sneakerCost
sneakers["Sneaker Cost"] = sneakerCost
sneakers = sneakers[["order_id", "shop_id", "user_id", "order_amount", "total_items", "Sneaker Cost", "payment_method",
                    "created_at"]]
sneakers
```

```python
AOV = sneakers["order_amount"].sum()/5000
AOV
```

#### below table verifies that the amount remains the same regardless of the payment method, as the max price and avg price are same.

```python
shops = pd.DataFrame(columns={"shop_id", "Max Price", "Avg Price"})
shops["shop_id"] = sneakers["shop_id"].unique()
shops = shops.sort_values(by="shop_id")
shops.reset_index(inplace=True)
shopAvg = sneakers.groupby(by="shop_id").mean()
shopAvg.reset_index(inplace=True)
shopMax = sneakers.groupby(by="shop_id").max()
shopMax.reset_index(inplace=True)
shops = shops.drop(columns={"index"})
for i in range(0, len(shops)):
    shops["Avg Price"][i] = shopAvg["Sneaker Cost"][i]
    shops["Max Price"][i] = shopMax["Sneaker Cost"][i]

shops
```

#### a) the value calculated originally is higher due to the price point set by store #78 of 25725 (a possible outlier). Which eventually is moving the average way up.

#### b) we could instead use modal value (mode), that will tell the consistent amount a consumer is paying throughout all these stores. 

```python
sneakers["order_amount"].mode()
```

```python
shops["Max Price"].mode()
```

#### c) here above the modal order amount is 153 is what seems reasonable as many orders have been made out around this amount and interestingly it is price that most stores have set for the sneakers!
